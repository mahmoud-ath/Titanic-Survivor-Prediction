Precision, Recall, and F1 Score Comparison

```python
from matplotlib import pyplot as plt
import numpy as np

# Assuming precision, recall, and f1 score are calculated for both models
precision_knn_scratch = 0.6875
recall_knn_scratch = 0.663
f1_knn_scratch = 0.669

precision_knn_sklearn = 0.80
recall_knn_sklearn = 0.71
f1_knn_sklearn = 0.75

# Plot the metrics for both models
labels = ['Precision', 'Recall', 'F1 Score']
knn_scratch_metrics = [precision_knn_scratch, recall_knn_scratch, f1_knn_scratch]
knn_sklearn_metrics = [precision_knn_sklearn, recall_knn_sklearn, f1_knn_sklearn]

x = np.arange(len(labels))  # Label locations

fig, ax = plt.subplots(figsize=(10, 6))
width = 0.3  # Bar width

# Plot bars for both models
ax.bar(x - width/2, knn_scratch_metrics, width, label='KNN from Scratch', color='gray')
ax.bar(x + width/2, knn_sklearn_metrics, width, label='KNN from Sklearn', color='green')

# Add labels and title
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Precision, Recall, and F1 Score Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.tight_layout()
plt.show()
```
